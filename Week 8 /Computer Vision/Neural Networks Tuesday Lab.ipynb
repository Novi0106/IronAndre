The author of **the article underlines that error rates in facial recognition are higher for any group outside of white males**. He ascribes this result to both the methodology and the training models/data being centered around white males and resulting in inaccurate results for anyone falling outside of this pattern.

He underlines his concerns that continuing down the same road would widen existing gaps and injustices affecting already marginalised groups. Hence, he advocates for a more equitable and conscious approach in composing relevant data sets. Aknowledging that it is not only a matter of methodology, he also calls for reform of wider issues such as bills to reform police work.
