```python
import docx2txt
import nltk
```

```python
nltk.download()
```

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text
```

```python
# define variable
# use nltk to separate the individual words - word_tokenize

textfile = docx2txt.process('Tokyo.docx')
```

```python
token = word_tokenize(textfile)
```

```python
token[:100]
```

```python
len(token)
```

```python
fdist1 = nltk.FreqDist(token)
```

```python
word_freq = dict((word, freq) for word, freq in fdist1.items() if not word.isdigit())
```

```python
word_freq
```

```python
# cleaning / wrangling steps:

# prepositions for / in / a / in - stop words
# pronouns - personal I / you / he / she might not adds meaning
# stemming - Plan, plans
# punctuation 
# lower case
```

```python
# get rid of punctuation
clean_words = [word for word in token if word.isalpha()]
```

```python
clean_words
```

```python
# make everything lower case
lower_words = [word.lower() for word in clean_words]
```

```python
lower_words
```

```python
# remove stop words

from nltk.corpus import stopwords

excluding_stop = stopwords.words('english')
```

```python
words_minus_stop = [word for word in lower_words if word not in excluding_stop ]
```

```python
len(words_minus_stop)
```

```python
fdist2 = nltk.FreqDist(words_minus_stop)
```

```python
fdist2.most_common(20)
```

```python
textlist = Text(words_minus_stop)
```

```python
# words that frequently appear together (phrases)
textlist.collocations()
```

```python
# find fragments containing key words 
# another index is created - so searching is faster next time

textlist.concordance('measures')
```

```python
# lexical dispersion plot - analyses appearances of words
textlist.dispersion_plot(['tokyo', 'japan', 'athletes', 'olympics', 'village', 'vaccinated'])
```
