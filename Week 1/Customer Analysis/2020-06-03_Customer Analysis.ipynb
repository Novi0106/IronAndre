**Adjusting the customer lifetime value column**
**This seems a bit of. The values in the initial files seemed way too high and a large bit was str-type.**
**Just looking at the instructions for the lab I would assumed they are float**
**Not really sure if my way is very elegant, since I converted all of them to strings first.**

```python
customer_data1['Customer Lifetime Value'] = customer_data1['Customer Lifetime Value'].astype('str', errors = 'ignore')
customer_data1['Customer Lifetime Value'] = list(map(lambda x : x.strip('%'), customer_data1['Customer Lifetime Value']))
```

```python
customer_data1['Customer Lifetime Value'] = pd.to_numeric(customer_data1['Customer Lifetime Value'], errors = "coerce")
```

```python
customer_data1['Customer Lifetime Value'] = customer_data1['Customer Lifetime Value'].apply(lambda x: x*100)
```
